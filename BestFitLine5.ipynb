```python
style.use('fivethirtyeight')



xs =np.array([100, 115, 122, 135, 140, 107, 116, 113, 118, 110, 127, 102], dtype=np.float64)
ys = np.array([1900, 2450, 2700, 2750, 2800, 2350, 2450, 2600, 2300, 2100, 2500, 2000], dtype=np.float64) 

def best_fit_slope(xs, ys):
    m = (((mean(xs) * mean(ys)) - mean(xs*ys))/
         (mean(xs) * mean(xs)) - mean (xs*xs))
    b = mean(ys) - m*mean(xs)
    return m, b

m, b = best_fit_slope(xs, ys)
print (m,b)
regression_line = [(m*x)+b for x in xs]

plt.scatter(xs, ys)
plt.plot(xs, regression_line)
plt.show()
```
